```python
import re
```

```python
input_file = "2025-06-20_09-09-01_SPHINCSplus.txt"
base_name = input_file.replace(".txt", "")

with open(input_file, 'r') as f:
    content = f.read()

parts = re.split(r'Launching experiment', content)[1:]
```

```python
print(len(parts))
```

```python
for i in range(len(parts)):
    exp_num = i + 1
    output_file = f"{base_name}_exp{exp_num}.txt"
    with open(output_file, 'w') as f:
        f.write(parts[i])
```
